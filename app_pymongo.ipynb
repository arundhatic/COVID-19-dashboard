```python
import json
import pymongo
import pprint
```

```python
client = MongoClient('localhost', 27017)
db = client['confirmed_db']
collection_confirmed = db['confirmed']
```

```python
json_file = './csse_covid_19_time_series/time_series_covid19_confirmed_global.json'
json_file1 = './csse_covid_19_time_series/time_series_covid19_deaths_global.json'
json_file2 = './csse_covid_19_time_series/time_series_covid19_recovered_global.json'
```

```python
with open(json_file) as f:
     data_json = json.load(f)
```

```python
# if pymongo >= 3.0 use insert_many() for inserting many documents
collection_confirmed.drop()
for data in data_json:
    collection_confirmed.insert_one(data)
   
```

```python
confirmed_data = collection_confirmed.find()
```

```python
for c in confirmed_data:
    print(c)
```

```python
db = client['deaths_db']
collection_deaths = db['deaths']
with open(json_file1) as f:
     data_json = json.load(f)
        
collection_deaths.drop()
for data in data_json:
    collection_deaths.insert_one(data)
    
deaths_data = collection_deaths.find()
for d in deaths_data:
    print(d)
```

```python
db = client['recovered_db']
collection_recovered = db['recovered']
with open(json_file2) as f:
     data_json = json.load(f)
        
collection_recovered.drop()
for data in data_json:
    collection_recovered.insert_one(data)
    
recovered_data = collection_recovered.find()
for r in recovered_data:
    print(r)
```
